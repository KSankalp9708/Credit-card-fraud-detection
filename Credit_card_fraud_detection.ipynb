```python
#Importing important libraries
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset to a Pandas dataframe
Credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# Listing first 5 rows of the dataset
Credit_card_data.head()
```

```python
# Listing last 5 rows of the dataset
Credit_card_data.tail()
```

```python
# Retrieving dataset information
Credit_card_data.info()
```

```python
# Checking the number of missing values in each column
Credit_card_data.isnull().sum()
```

Thus, there is **no** any missing value in any of the columns in the dataset.

```python
# distribution of legit transactions & fraud transactions
Credit_card_data['Class'].value_counts()
```

In the above table, 0 represents normal transactions, while 1 represents fraudulent transations.

As the given data is highly unbalanced, we need to separate the two classes of transactions.

```python
# Separating the datasets into classes: 'legit' and 'fraud' for analysis:
legit = Credit_card_data[Credit_card_data.Class == 0]
fraud = Credit_card_data[Credit_card_data.Class == 1]
```

```python
#checking shape(rows, columns) for both class:
print(legit.shape)
print(fraud.shape)
```

```python
# Finding out statistical measures of the data:
#First, for legit class:
legit.Amount.describe()
```

```python
#Similarly, for fraud class:
fraud.Amount.describe()
```

```python
# Comparing the values for both transactions
Credit_card_data.groupby('Class').mean()
```

**Under-Sampling**

To build a sample dataset containing similar distribution of normal transactions and fraud transactions

As number of fraud transactions = 492

For similar distribution, taking a sample of 492 legit transactions.

```python
legit_sample = legit.sample(n=492)
```

*Concatenating* two DataFrames

```python
New_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
# Listing first 5 rows of the new dataset
New_dataset.head()
```

```python
# Listing the last 5 rows of the new dataset
New_dataset.tail()
```

```python
#Checking the no. of rows of fraud and legit transactions in the new dataset.
#Both should be same and =492.
New_dataset['Class'].value_counts()
```

```python
# Again, comparing the values for both transactions for this new dataset
New_dataset.groupby('Class').mean()
```

Splitting the data into Features(X) & Targets(Y)

```python
X = New_dataset.drop(columns='Class', axis=1)
Y = New_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**

*Logistic Regression*

```python
Model = LogisticRegression()
```

```python
# Training the Logistic Regression Model with the training data
Model.fit(X_train, Y_train)
```

**Model Evaluation**

*Accuracy Score*

```python
# Determining accuracy on training data
X_train_prediction = Model.predict(X_train)
Training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on the training data : ', Training_data_accuracy)
```
