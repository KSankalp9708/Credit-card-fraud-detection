Importing important libraries

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

 Loading the dataset to a dataframe

```python
Credit_card_data = pd.read_csv('/content/creditcard.csv')
```

Data Reading & Cleaning

```python
# Listing first 5 rows of the dataset
Credit_card_data.head()
```

```python
# Listing last 5 rows of the dataset
Credit_card_data.tail()
```

```python
# Retrieving dataset information
Credit_card_data.info()
```

```python
# Checking the number of missing values in each column
Credit_card_data.isnull().sum()
```

Thus, there are no any data/ column entry that is missing.

```python
# distribution of legit transactions & fraud transactions
Credit_card_data['Class'].value_counts()
```

In the above table, 0 represents normal transactions, while 1 represents fraudulent transations.

As the given data is highly unbalanced, we need to separate the two classes of transactions.

```python
# Separating the datasets into classes: 'legit' and 'fraud' for analysis:
legit = Credit_card_data[Credit_card_data.Class == 0]
fraud = Credit_card_data[Credit_card_data.Class == 1]
```

```python
#checking shape(rows, columns) for both class:
print(legit.shape)
print(fraud.shape)
```

```python
# Finding out statistical measures of the data:
#First, for legit class:
legit.Amount.describe()
```

```python
#Similarly, for fraud class:
fraud.Amount.describe()
```

```python
# Comparing the values for both transactions
Credit_card_data.groupby('Class').mean()
```

**Under-Sampling**

To build a sample dataset containing similar distribution of normal transactions and fraud transactions

As number of fraud transactions = 492

For similar distribution, taking a sample of 492 legit transactions.

```python
legit_sample = legit.sample(n=492)
```

*Concatenating* two DataFrames

```python
New_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
# Listing first 5 rows of the new dataset
New_dataset.head()
```

```python
# Listing the last 5 rows of the new dataset
New_dataset.tail()
```

```python
#Checking the no. of rows of fraud and legit transactions in the new dataset.
#Both should be same and =492.
New_dataset['Class'].value_counts()
```

```python
# Again, comparing the values for both transactions for this new dataset
New_dataset.groupby('Class').mean()
```

Splitting the data into Features(X) & Targets(Y)

```python
X = New_dataset.drop(columns='Class', axis=1)
Y = New_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training & Evaluation**

```python
#Initializing models
Models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}
```

```python
# Training the models
# Then, predicting using X_test dataset and finally calculating accuracy

results = []

for name, model in Models.items():
    model.fit(X_train,Y_train)
    Y_pred = model.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_pred)
    precision = precision_score(Y_test, Y_pred)
    recall = recall_score(Y_test, Y_pred)
    f1 = f1_score(Y_test, Y_pred)
    results.append({
        "Model": name,
        "Accuracy": round(accuracy * 100, 2),
        "F1 Score": round(f1 * 100, 2),
        "Precision": round(precision * 100, 2),
        "Recall": round(recall * 100, 2)
      })
```

*Results Table*

```python
# Display results as a DataFrame
df_results = pd.DataFrame(results)
print(df_results)
```
